```python
import matplotlib.pyplot as plt

city = ["Delhi", "Mumbai", "Kolkata", "Chennai","Rajasthan"]
population = [32900000, 21700000, 15100000, 11300000,10101010]
plt.figure(figsize=(10,5),facecolor='purple') #size of L*B
plt.plot(city, population,marker='o',color='black',alpha=0.8,linewidth=3,markeredgewidth=4)
plt.xlabel("city")
plt.ylabel("population")
plt.gca().set_facecolor('green') #gca for the inner background of the graph
plt.title("City wise Population Dataset") #for the title
plt.grid()#for the grid in the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

city = ["Delhi", "Mumbai", "Kolkata", "Chennai", "Rajasthan"]
population = [32900000, 21700000, 15100000, 11300000, 10101010]

plt.figure(figsize=(10, 10), facecolor='purple')

bar_colors = ['red', 'blue', 'orange', 'yellow', 'purple']
plt.barh(city, population, color=bar_colors)
plt.xlabel("population")
plt.ylabel("city")
plt.gca().set_facecolor('green')
plt.title("City wise Population Dataset")
plt.grid(axis='x') # Grid lines on the x-axis make more sense for horizontal bars
plt.show()
```

```python
# Histogram: Shows distribution of population values
plt.figure(figsize=(10,5), facecolor='purple')
plt.hist(population, bins=3, color='orange', edgecolor='black')
plt.title("Population Distribution")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.gca().set_facecolor('green')
plt.grid()
plt.show()
```

```python
# Pie Chart: Shows proportions of the whole
plt.figure(figsize=(8, 8))
plt.pie(population, labels=city, autopct='%1.1f%%', colors=['red', 'blue', 'orange', 'yellow', 'purple'])
plt.title("City wise Population Proportions")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(population,city)
plt.title("City wise Population Proportions")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.boxplot(population, labels=city, autopct='%1.1f%%', colors=['red', 'blue', 'orange', 'yellow', 'purple'])
plt.title("City wise Population Proportions")
plt.show()
```
