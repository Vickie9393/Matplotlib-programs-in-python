```python
# 1. Filter for three courses
selected_courses = ["B.Tech", "BCA", "MCA"]
filtered_df = df[df['Course'].isin(selected_courses)]
pivot_df = filtered_df.pivot(index='Student_ID', columns='Course', values='Marks').fillna(0)

# 3. Create the stack plot
plt.figure(figsize=(10, 6))
plt.stackplot(pivot_df.index, 
              pivot_df['B.Tech'], 
              pivot_df['BCA'], 
              pivot_df['MCA'], 
              labels=['B.Tech', 'BCA', 'MCA'], 
              alpha=0.8)

# Add titles, labels, and legend
plt.title('Stack Plot of Marks by Course')
plt.xlabel('Student ID')
plt.ylabel('Marks')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# 1. Group by Age and calculate mean marks, then sort by Age
avg_marks_by_age = df.groupby('Age')['Marks'].mean().sort_index()

# 2. Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(avg_marks_by_age.index, avg_marks_by_age.values, marker='o', linestyle='-', color='teal')

# Add titles and labels
plt.title('Average Marks by Age')
plt.xlabel('Age')
plt.ylabel('Average Marks')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#Group by Course and calculate mean and max marks
avg_marks_by_course = df.groupby('Course')['Marks'].mean()
max_marks_by_course = df.groupby('Course')['Marks'].max()

plt.figure(figsize=(10, 6))

plt.plot(avg_marks_by_course.index, avg_marks_by_course.values, 
         marker='o', linestyle='-', color='blue', label='Average Marks')

plt.plot(max_marks_by_course.index, max_marks_by_course.values, 
         marker='s', linestyle='--', color='red', label='Maximum Marks')

# 3. Add titles, labels, and legend
plt.title('Average vs Maximum Marks by Course')
plt.xlabel('Course')
plt.ylabel('Marks')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#  Sort by Marks and get Top 10
top_10 = df.sort_values(by='Marks', ascending=False).head(10)

colors = ['green' if m >= 85 else 'red' for m in top_10['Marks']]

# 3. Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10['Name'], top_10['Marks'], color=colors, edgecolor='black')

plt.title('Top 10 Students by Marks')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Create a grouped bar chart comparing:Average Marks Maximum Marks for each Course.
import matplotlib.pyplot as plt
import numpy as np

avg_marks = df.groupby('Course')['Marks'].mean()
max_marks = df.groupby('Course')['Marks'].max()

# 2. Setup positions
courses = avg_marks.index
x = np.arange(len(courses)) 
width = 0.35 

# 3. Create the grouped bar chart
plt.figure(figsize=(10, 6))

# Use x - width/2 for average and x + width/2 for max
plt.bar(x - width/2, avg_marks, width, label='Average Marks', color='skyblue')
plt.bar(x + width/2, max_marks, width, label='Maximum Marks', color='orange')

# 4. Final touches
plt.title('Average vs Maximum Marks by Course')
plt.xlabel('Course')
plt.ylabel('Marks')
plt.xticks(x, courses)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
